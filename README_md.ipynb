Array adalah struktur data yang digunakan untuk menyimpan banyak nilai dalam satu variabel dengan tipe data yang sama.

Pada program tersebut, array digunakan dengan list Python:
nilai = []
Artinya kita membuat array kosong bernama nilai untuk menyimpan nilai mahasiswa.

Kemudian nilai dimasukkan dengan:
nilai.append(n)
Setiap input nilai akan ditambahkan ke dalam array nilai, sehingga semua nilai mahasiswa tersimpan dalam satu tempat.

Kesimpulan singkat:
Array pada program ini berfungsi untuk menyimpan 10 nilai mahasiswa dalam satu variabel agar mudah dihitung nilai tertinggi, terendah, rata-rata, dan jumlah kelulusan.

```python
import matplotlib.pyplot as plt

# array untuk menyimpan nilai
nilai = []

# input nilai
for i in range(10):
    n = int(input(f"Masukkan nilai mahasiswa ke-{i+1}: "))
    nilai.append(n)

# nilai tertinggi dan terendah
nilai_tertinggi = max(nilai)
nilai_terendah = min(nilai)

# rata-rata
rata_rata = sum(nilai) / len(nilai)
# hitung kelulusan
lulus = 0
tidak_lulus = 0

for n in nilai:
    if n >= 60:
        lulus += 1
    else:
        tidak_lulus += 1

# output
print("\nDaftar Nilai:", nilai)
print("Nilai Tertinggi:", nilai_tertinggi)
print("Nilai Terendah:", nilai_terendah)
print("Rata-rata:", rata_rata)
print("Jumlah Lulus:", lulus)
print("Jumlah Tidak Lulus:", tidak_lulus)

# grafik nilai tertinggi dan terendah
plt.figure(figsize=(5,4))
plt.bar(["Tertinggi", "Terendah"], [nilai_tertinggi, nilai_terendah], color=["green","red"])
plt.title("Grafik Nilai Tertinggi & Terendah")
plt.ylabel("Nilai")
plt.show()# grafik kelulusan
plt.figure(figsize=(5,4))
plt.pie([lulus, tidak_lulus], labels=["Lulus","Tidak Lulus"], autopct="%1.1f%%", colors=["blue","orange"])
plt.title("Grafik Kelulusan Mahasiswa")
plt.show()
```
