```python
import numpy as np
arr = np.array([10,20,30,40,50])
print(arr)
```

nilai tertinggi dalam array

```python
nilai_tertinggi = np.max(arr)
print(nilai_tertinggi)
```

nilai terendah dalam array

```python
nilai_terendah = np.min(arr)
print(nilai_terendah)
```

nilai rata-rata dalam array

```python
nilai_rata = np.mean(arr)
print(nilai_rata)
```
