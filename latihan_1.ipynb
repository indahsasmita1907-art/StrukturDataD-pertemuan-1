```
mahasiswa = ["Andi","Budi","Citra","Devi","Eko"]
print("List Mahasiswa: ")
print(mahasiswa)
```

```
mahasiswa.append("Fajar")
print("setelah ditambahkan nama baru")
print(mahasiswa)
```

```
mahasiswa.remove("Citra")
print("setelah mengapus nama Citra ")
print(mahasiswa)
```
